```
import pandas as pd
dfk = pd.DataFrame()
dfp = pd.DataFrame()
dfc = pd.DataFrame()


import zipfile
with zipfile.ZipFile('./fake-and-real-news-dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('./')
fake=pd.read_csv('Fake.csv')
category=['FALSE']*(23481)
fake['Category']=category
fake.head() 
true=pd.read_csv('True.csv')
category=['TRUE']*(21417)
true['Category']=category
true.at[20, 'text']  
kaggle=true.head(5000).append(fake.head(5000),ignore_index=True)

import zipfile
with zipfile.ZipFile('./covid19.zip', 'r') as zip_ref:
    zip_ref.extractall('./')
    import zipfile
with zipfile.ZipFile('./politifact.zip', 'r') as zip_ref:
    zip_ref.extractall('./')
fake.at[23, 'text']
politi = pd.read_csv('Politifact_.csv')

covid = pd.read_csv('COVID19FN.csv')
covid.at[2818, 'Text']
```

```
kaggle
dfk['text'] = kaggle['text']
dfk['Category'] = kaggle['Category'] 
```

```
politi
dfp['text'] = politi['text']
dfp['Category'] = politi['is_fake']  
for i in range(len(dfp)):
  if(dfp.at[i, 'Category']=="TRUE" or dfp.at[i, 'Category']=='True'):
    dfp.at[i, 'Category']=0
  elif(dfp.at[i, 'Category']=="FALSE" or dfp.at[i, 'Category']=='False'):
    dfp.at[i, 'Category']=1
dfp['Category']=dfp['Category'].astype('int') 
```

```
covid
dfc['text'] = covid['text']
dfc['Category'] = covid['Label'] 
```

```
dfk=dfk.sample(frac=1).reset_index(drop=True)
print('kaggle')
print(dfk.groupby(['Category']).size())
dfp=dfp.sample(frac=1).reset_index(drop=True)
print('poitifact')
print(dfp.groupby(['Category']).size())
dfc=dfc.sample(frac=1).reset_index(drop=True)
print('covid19')
print(dfc.groupby(['Category']).size())
```

```
df = pd.DataFrame()

df = df.append(dfc)
df = df.append(dfk)
df = df.sample(frac=1).reset_index(drop=True)
```

```

# df.iloc[i, 'Category']
df.at[0, 'Category']
for i in range(len(df)):
  if(df.at[i, 'Category']=="TRUE" or df.at[i, 'Category']=='True'):
    df.at[i, 'Category']=1
  elif(df.at[i, 'Category']=="FALSE" or df.at[i, 'Category']=='False'):
    df.at[i, 'Category']=0
df['Category']=df['Category'].astype('int')
```
